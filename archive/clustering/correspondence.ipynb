```python
import prince
import pandas as pd

df20 = pd.read_csv('../data/processed/df_2020_ohe.csv', usecols = ['service type', 'income'])
df21 = pd.read_csv('../data/processed/df_2021_ohe.csv')
```

```python
df20 = df20[df20['income'].isin(['Under $125,000', '$125,000 to under $2 million'])]
df20.dropna(inplace = True)
```

```python
mca = prince.MCA(
    n_components=2,
    n_iter=10,
    copy=True,
    check_input=True,
    engine='auto',
    random_state=42
)
```

```python
df20.describe()
```

```python
mca = mca.fit(df20)
```

```python
ax = mca.plot_coordinates(
    X=df20,
    ax=None,
    figsize=(20, 20),
    show_row_points=False,
    # row_points_size=10,
    show_row_labels=False,
    show_column_points=True,
    column_points_size=30,
    show_column_labels=True,
    legend_n_cols=1)
```
