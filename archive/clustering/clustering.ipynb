```python

import pandas as pd

df20 = pd.read_csv('../data/processed/df_2020_ohe.csv', usecols = [ 'service type','paid staff','volunteers', 'income', 'service delivery affect','service delivery change'])
```

```python
df20.dropna(inplace=True)
```

```python
df20.describe()
```

```python
import gower
import numpy as np
```

```python
X = np.asarray(df20)
```

```python
distance_matrix = gower.gower_matrix(df20)
```

```python
from sklearn.cluster import DBSCAN

# Configuring the parameters of the clustering algorithm
dbscan_cluster = DBSCAN(eps=0.3, 
                        min_samples=2, 
                        metric="precomputed")

# Fitting the clustering algorithm
dbscan_cluster.fit(distance_matrix)

# Adding the results to a new column in the dataframe
df20["cluster"] = dbscan_cluster.labels_
```
